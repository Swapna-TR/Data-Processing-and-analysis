Problem 1:Employee Performance Bonus Eligibility

```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# first I will find the highest score manually
highest = 0

for emp in employees:
    if employees[emp] > highest:
        highest = employees[emp]

# now check who got that highest score
top_list = []

for emp in employees:
    if employees[emp] == highest:
        top_list.append(emp)

print("Top Performers Eligible for Bonus:", end=" ")
for name in top_list:
    print(name, end=" ")
print("(Score:", highest, ")")
```

Problem 2:Search Query Keyword Analysis

```python
sentence = "Buy mobile phone buy phone online"

# convert to lowercase so words match
sentence = sentence.lower()

# remove punctuation one by one (simple way)
sentence = sentence.replace(".", "")
sentence = sentence.replace(",", "")

word_list = sentence.split()

count_dict = {}

# count words using simple check
for word in word_list:
    if word not in count_dict:
        count_dict[word] = 1
    else:
        count_dict[word] = count_dict[word] + 1

# print only words that appear more than once
result = {}

for word in count_dict:
    if count_dict[word] > 1:
        result[word] = count_dict[word]

print("Repeated words:", result)
```

Problem 3:Sensor Data Validation

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_pairs = []

# check each reading with index
i = 0

for value in sensor_readings:
    if value % 2 == 0:   # even means valid
        valid_pairs.append((i, value))
    i = i + 1

print("Valid Sensor Readings:", valid_pairs)
```

Problem 4: Email Domain Usage Analysis

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_data = {}

# extract domain and count
for mail in emails:
    parts = mail.split("@")
    domain = parts[1]

    if domain not in domain_data:
        domain_data[domain] = 1
    else:
        domain_data[domain] = domain_data[domain] + 1

total = len(emails)

print("Domain usage percentage:")

for domain in domain_data:
    percent = (domain_data[domain] / total) * 100
    print(domain, "-", int(percent), "%")
```

Problem 5:Sales Spike Detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

total_sales = 0

# calculate total first
for s in sales:
    total_sales = total_sales + s

avg = total_sales / len(sales)

# spike level = 30% more than average
spike_level = avg + (avg * 0.30)

print("Spike days are:")

day = 1
for s in sales:
    if s > spike_level:
        print("Day", day, ":", s)
    day = day + 1
```

Problem 6:Duplicate User ID Detection

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

id_data = {}

# count each id
for uid in user_ids:
    if uid not in id_data:
        id_data[uid] = 1
    else:
        id_data[uid] = id_data[uid] + 1

print("Duplicate IDs found:")

for uid in id_data:
    if id_data[uid] > 1:
        print(uid, "appears", id_data[uid], "times")
```
